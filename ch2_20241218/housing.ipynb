## 모델 학습 및 평가

### train, test 데이터셋 분할

```python
from sklearn.model_selection import train_test_split

#독립 변수와 종속 변수 나누기
X = df.drop('MEDV', axis=1)
y = df['MEDV']

#전체 데이터 중 20%를 테스트 데이터로 분할, 무작위로 섞어서 추출
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True, random_state=42)  #shuffle=True,

#분할된 데이터 모양 확인
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
#학습 데이터 확인
X_train
```

### 선형 회귀 모델

#### 스케일링
Linear Regression과 같은 선형 모델에서 사용
트리 기반 모델은 특성의 값을 직접 비교하지 않고 특정 기준값을 사용하여 데이터를 분할하기 때문에 스케일 조정이 필요하지 않음

```python
from sklearn.preprocessing import StandardScaler

#데이터 스케일링
scaler = StandardScaler()   #데이터의 평균을 0, 분산을 1로 스케일링
scale_X_train = scaler.fit_transform(X_train)   #훈련 데이터셋 스케일링하고 변환
scale_X_test = scaler.transform(X_test)         #테스트 데이터셋 변환
```

```python
scale_X_train[0]
```

#### 모델 학습

```python
from sklearn.linear_model import LinearRegression

#선형 회귀 모델 생성
linear_regression_model = LinearRegression()    
linear_regression_model.fit(scale_X_train, y_train)         #모델 학습

lr_pred = linear_regression_model.predict(scale_X_test)     #예측
```

#### 모델 평가

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

lr_mae = mean_absolute_error(y_test, lr_pred)                   #mae 평가
lr_mse = mean_squared_error(y_test, lr_pred)                    #mse 평가
lr_rmse = mean_squared_error(y_test, lr_pred, squared=False)    #mse에 제곱근을 취한 rmse 평가
lr_r2 = r2_score(y_test, lr_pred)                               #r2 평가

print("mean absolute error:", lr_mae)
print("mean squared error:", lr_mse)
print("Root mean squared error:", lr_rmse)
print("r2 score:", lr_r2)
```

### 예측 값과 실제 값 비교 시각화

```python
#가장 상관계수가 높은 RM 컬럼의 실제 값과 예측 값을 함께 산점도로 그려 비교하는 함수
def real_predict_plot_scatter(predict_value):

    plt.figure(figsize=(7, 5))      #그래프 크기 조절
    plt.scatter(X_test['RM'], y_test, label='real_value')               #RM 컬럼의 실제 값 표시
    plt.scatter(X_test['RM'], predict_value, label='predict_value')     #RM 컬럼의 예측 값 표시

    plt.title('Real vs Predicted Values for RM Feature')                #그래프 제목 표시
    plt.legend(loc='best')          #최적의 위치에 범례 표시

    plt.show()                      #그래프 그리기
```

```python
#RM 컬럼의 실제 값과 예측 값을 산점도로 그려 비교
real_predict_plot_scatter(lr_pred)
```

### 의사결정나무

```python
from sklearn.tree import DecisionTreeRegressor

#의사결정나무 모델 생성
decision_tree_model = DecisionTreeRegressor(random_state=42)    
decision_tree_model.fit(X_train, y_train)                       #모델 학습

dt_pred = decision_tree_model.predict(X_test)                   #예측

dt_mae = mean_absolute_error(y_test, dt_pred)                   #mae 평가
dt_mse = mean_squared_error(y_test, dt_pred)                    #mse 평가
dt_rmse = mean_squared_error(y_test, dt_pred, squared=False)    #mse에 제곱근을 취한 rmse 평가
dt_r2 = r2_score(y_test, dt_pred)                               #r2 평가

print("mean absolute error:", dt_mae)
print("mean squared error:", dt_mse)
print("Root mean squared error:", dt_rmse)
print("r2 score:", dt_r2)
```

### 랜덤포레스트 모델

```python
from sklearn.ensemble import RandomForestRegressor

#랜덤포레스트 회귀 모델 생성 (트리 수 100)
random_forest_model = RandomForestRegressor(random_state=42, n_estimators=100)  
random_forest_model.fit(X_train, y_train)                       #모델 학습

rf_pred = random_forest_model.predict(X_test)                   #예측

rf_mae = mean_absolute_error(y_test, rf_pred)                   #mae 평가
rf_mse = mean_squared_error(y_test, rf_pred)                    #mse 평가
rf_rmse = mean_squared_error(y_test, rf_pred, squared=False)    #rmse 평가
rf_r2 = r2_score(y_test, rf_pred)                               #r2 평가

print("mean absolute error:", rf_mae)
print("mean squared error:", rf_mse)
print("Root mean squared error:", rf_rmse)
print("r2 score:", rf_r2)
```

```python
#RM 컬럼의 실제 값과 예측 값을 산점도로 그려 비교
real_predict_plot_scatter(rf_pred)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#feature 중요도 추출
feature_importances = random_forest_model.feature_importances_

#feature와 중요도를 데이터프레임으로 변환
feature_importances_df = pd.DataFrame({
    'Feature': X.columns,
    'Importance': feature_importances
})

#중요도 순으로 정렬
feature_importances_df = feature_importances_df.sort_values(by='Importance', ascending=False)

#시각화
plt.figure(figsize=(5, 3))
sns.barplot(x='Importance', y='Feature', data=feature_importances_df)
plt.title('Feature Importances in Random Forest')
plt.show()
```

```python
import xgboost as xgb
from sklearn.metrics import mean_squared_error

#XGBoost 모델 생성 (트리 개수 100, 이건 결과 반영 0.2, 트리 깊이 3)
xgb_model = xgb.XGBRegressor(n_estimators=100, learning_rate=0.2, max_depth=3, random_state=42)
xgb_model.fit(X_train, y_train)                                     #모델 학습

xgb_pred = xgb_model.predict(X_test)                                #예측

# 평가
xgb_mae = mean_absolute_error(y_test, xgb_pred)                     #mae 평가
xgb_mse = mean_squared_error(y_test, xgb_pred)                      #mse 평가
xgb_rmse = mean_squared_error(y_test, xgb_pred, squared=False)      #rmse 평가
xgb_r2 = r2_score(y_test, xgb_pred)                                 #r2 평가

print("mean absolute error:", xgb_mae)
print("mean squared error:", xgb_mse)
print("Root mean squared error:", xgb_rmse)
print("r2 score:", xgb_r2)
```

```python
#RM 컬럼의 실제 값과 예측 값을 산점도로 그려 비교
real_predict_plot_scatter(xgb_pred)
```

### 모델 성능 평가

```python
#각 모델의 mae, mse, r2 평가 결과 출력 함수
def print_result(v):
    print(f"Mean Absolute Error: {v[0]:.2f}")
    print(f"Mean Squared Error: {v[1]:.2f}")
    print(f"R2 Score: {v[2]:.2f}")
```

```python
#모델 별 평가 결과를 모은 리스트
lr_results = [lr_mae, lr_mse, lr_r2]
df_results = [dt_mae, dt_mse, dt_r2]
rf_results = [rf_mae, rf_mse, rf_r2]
xgb_results = [xgb_mae, xgb_mse, xgb_r2]

#평가 결과 출력
print("----Linear Regression----")
print_result(lr_results)

print("\n------Decision Tree------")
print_result(df_results)

print("\n------Random Forest------")
print_result(rf_results)

print("\n---------XGBoost---------")
print_result(xgb_results)
```

## 결과 분석

```python
#mae, rmse, r2 평가 별 모델 결과 리스트
mae_results = [lr_mae, dt_mae, rf_mae, xgb_mae]
rmse_results = [lr_rmse, dt_rmse, rf_rmse, xgb_rmse]
r2_results = [lr_r2, dt_r2, rf_r2, xgb_r2]
```

```python
#모델의 성능 결과 막대 그래프로 그리기
models=['lr', 'dt', 'rf', 'xgb']        #모델 리스트
metrics = ['mae', 'rmse', 'r2']         #평가 리스트
results = [mae_results, rmse_results, r2_results]    #평가 별 모델 결과가 저장된 리스트

color=['C7', 'C7', 'C7','#e35f62']      #좋은 성능의 모델을 빨간색으로 표시하기 위한 색깔 저장 리스트

plt.figure(figsize=(10,3))              #그래프 크기 지정

for i, metrics in enumerate(metrics):
    plt.subplot(1, 3, i+1)              #그래프의 위치 지정
    plt.title(metrics.upper())          #그래프 제목 표시
    
    x_data = models                     #x값: 모델
    y_data = results[i]                 #y값: 평가 결과

    plt.bar(x_data,y_data, color=color, width=0.5)  #막대 그래프 그리기
```

```python
#예측 결과 시각화
plt.figure(figsize=(15, 5))     #그래프 크기 지정

#모델 이름과 예측 값을 저장한 리스트들
models = ['Linear Regression', 'Decision Tree Predictions', 'Random Forest Predictions', 'XGBoost']
preds = [lr_pred, dt_pred, rf_pred, xgb_pred]

for i in range(0, 4):
    plt.subplot(1, 4, i+1)                      #그래프의 위치 지정
    plt.scatter(y_test, preds[i], alpha=0.7)    #실제 값과 에측 값을 산점도로 표시
    plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')     #실제 값의 최소, 최대 값을 x, y축 범위로 설정, 기준 선을 빨간색으로 표시
    plt.title(models[i] + " Predictions")       #그래프 제목 표시
    plt.xlabel("Real Values")                   #x축 제목 표시
    plt.ylabel("Predicted Values ")             #y축 제목 표시
```
